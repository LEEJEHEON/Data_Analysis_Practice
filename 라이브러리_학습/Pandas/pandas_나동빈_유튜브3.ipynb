```python
# df의 마스킹

import numpy as np 
import pandas as pd 

df = pd.DataFrame(np.random.randint(1,10,(2,2)) , index=[0,1], columns=['A','B'])

print(df)

# 컬럼 A의 각 원소가 5보다 작거나 같은지 출력
print(df['A'] <= 5)

# 컬럼 A의 원소가 5보다 작고, 컬럼 B의 원소가 8보다 작은 행 추출
print(df.query('A < 5 & B < 8'))
```

```python
# df의 개별 연산 1

df = pd.DataFrame([[1,2,3,4],[1,2,3,4]], index = [0,1], columns = ['A','B','C','D'])
print (df)

df = df.apply(lambda x:x+1)
print(df)

def add_one(x):
    return x + 1

df = df.apply(add_one)
print(df)
```

```python
# df의 개별 연산 2

df = pd.DataFrame([['Apple','Apple','Carrot','Banana'],['Durian','Banana','Apple', 'Carrot']], index = [0,1], columns = ['A','B','C','D'])
print(df)

df = df.replace('Apple','Airport')
print(df)

df = df.replace({'Banana':'Ball'})
print(df)
```

```python
# df 의 그룹화 1 

df = pd.DataFrame([
    ['Apple', 7, 'Frult'],
    ['Banana', 3, 'Frult'],
    ['Beef', 5, 'Meal'],
    ['Kimchi', 4, 'Meal']] 
    , columns = ['Name', 'Frequency', 'Type']
    )

print(df)

print(df.groupby('Type').sum()) # 그룹화 후 합계
print(df.groupby('Type').mean('Frequency')) # 그룹화 후 평균
```

```python
# df 의 그룹화 2
import numpy as np

df = pd.DataFrame([
    ['Apple', 7,5, 'Frult'],
    ['Banana', 3,6, 'Frult'],
    ['Beef', 5,2, 'Meal'],
    ['Kimchi', 4,8, 'Meal']] 
    , columns = ['Name', 'Frequency','Importance', 'Type']
    )

print(df)
# print(df.groupby('Type').aggregate([min, max, np.average])) # 그룹화 후 여러 연산 적용 (np.average 에서 오류 발생)

print(df.groupby('Type').agg({'Frequency':'sum', 'Importance':'mean'})) # 그룹화 후 여러 연산 적용
```

```python
# df 의 그룹화 3 

df = pd.DataFrame([
    ['Apple', 7,5, 'Frult'],
    ['Banana', 3,6, 'Frult'],
    ['Beef', 5,2, 'Meal'],
    ['Kimchi', 4,8, 'Meal']] 
    , columns = ['Name', 'Frequency','Importance', 'Type']
    )

def my_filter(x):
    return x['Frequency'].mean() >= 5

print(df)
df = df.groupby('Type').filter(my_filter)
print(df)
```

```python
# df 의 그룹화 4

df = pd.DataFrame([
    ['Apple', 7,5, 'Frult'],
    ['Banana', 3,6, 'Frult'],
    ['Beef', 5,2, 'Meal'],
    ['Kimchi', 4,8, 'Meal']] 
    , columns = ['Name', 'Frequency','Importance', 'Type']
    )

df = df.groupby('Type').get_group('Frult')

print (df)
```

```python
# df 의 그룹화 5

df = pd.DataFrame([
    ['Apple', 7,5, 'Frult'],
    ['Banana', 3,6, 'Frult'],
    ['Beef', 5,2, 'Meal'],
    ['Kimchi', 4,8, 'Meal']] 
    , columns = ['Name', 'Frequency','Importance', 'Type']
    )

df['Gap'] = df.groupby('Type')['Frequency'].transform(lambda x: x - x.mean())
print(df)


# df['Gap'] = df.groupby('Type')['Frequency'].apply(lambda x: x - x.mean()) # apply 에서 오류 발생 
# print(df)

```

```python
# df의 다중화 

df = pd.DataFrame(
    np.random.randint(1,10,(4,4)),
    index = [['1차','1차','2차','2차'], ['공격','수비','공격','수비']],
    columns = ['1회','2회','3회','4회']
)

print(df)
print(df[['1회','2회']].loc['2차'])
```

```python
# 피벗테이블의 기초 

df = pd.DataFrame([
    ['Apple', 7,5, 'Frult'],
    ['Banana', 3,6, 'Frult'],
    ['Beef', 5,2, 'Meal'],
    ['Kimchi', 4,8, 'Meal']] 
    , columns = ['Name', 'Frequency','Importance', 'Type']
    )

print(df)
df = df.pivot_table(index='Importance', columns ='Type', values='Frequency', aggfunc=np.max)

print(df)
```
